## Research Question Hypthesis
### Research question: 

We will examine whether professional critics, and general audiences evaluated movies released on streaming platforms (e.g., Netflix, Disney+) differently before (2018, 2019) and during the height of (2020, 2021) the COVID-19 pandemic, using data from Rotten Tomatoes and IMDb for films available between 2018 and 2021. 

1) What is the impact of the COVID-19 pandemic on changing audience/viewer sentiment towards all movies measured by average IMDb rating? 

2) What proportion of movies had higher Rotten Tomatoes critic scores (Tomatometer) than audience scores (Popcornmeter) in the pre-pandemic (2018-2019) and peak pandemic (2020-2021) periods? 

3) How did the COVID-19 pandemic affect the weighted-average rating patterns across different maturity ratings of media? More specifically, is there a statistically significant change between Parental-Guided rated movies (PG-13 + TV-PG + TV-14) vs Mature / Adult rated movies (R + TV-MA)?

### Evaluation Method/ Methodology:
We will evaluate our data using a comparative approach. We will split the dataset into the pre-pandemic (2018-2019) and peak-pandemic (2020-2021) periods and compute summary statistics for IMDb ratings, Rotten Tomatoes critic scores, and audience scores.
To test for statistically significant changes in average IMDb ratings across periods (Question 1), we will use two-sample t-tests.
We will then compare the proportion of films (Question 2) where the Tomatometer exceeds the Popcornmeter, which are the critics and audience scores, respectively, in each period using difference-in-proportions tests.
Lastly, we will calculate vote-weighted average ratings and use two-sample t-tests within each age-rating category to test for statistically significant differences between the two periods.
4. Hypothesis:

1) There is a statistically significant increase in the average IMDb rating for movies and TV shows released between the height of COVID-19 (2020, 2021) and 2018-2019, likely due to increased consumption and greater access to entertainment, with algorithms curated to users' tastes. 
2) The gap between Rotten Tomatoes’ Tomatometer (Critics) and Popcornmeter (Audience) scores widened significantly during the pandemic (2020-2021) compared to the pre-pandemic baseline, as critics focused on quality, while the general audience consumed more content to pass the time. 
3) During the pandemic, mature-rated media (R and TV-MA) showed an increase in total vote volume and a more consistent weighted-average rating than Parental-Guidance-rated media (PG-13, TV-PG, and TV-14). This divergence is primarily attributable to the transition in entertainment consumption from conventional cinema to digital streaming platforms (e.g., Netflix, Disney+).
