# **Reading in the datasets (the CSV files)**

```python
# Importing packages
import pandas as pd
import numpy as np


# Reading in all seven datasets. Some were previously .csv already but others need to be converted
netflix_1 = pd.read_csv("NetFlix.csv")
netflix_2 = pd.read_csv("netflix_titles.csv")
disney = pd.read_csv("disney_plus_titles.csv")
rotten_tomatoes_movies = pd.read_csv("rotten_tomatoes_movies.csv")
rotten_tomatoes_movies_reviews = pd.read_csv("rotten_tomatoes_movie_reviews.txt")
imdb_title_ratings = pd.read_csv("title.ratings.tsv", sep="\t")
imdb_title_basics = pd.read_csv("title.basics.txt", sep="\t")
```

## Netflix 1 Dataset

```python
netflix_1.head()
```

```python
print("There are", len(netflix_1), "rows in this dataset before cleaning.")
```

```python
# Replace missing values with None so Oracle can store it as NULL
netflix_1 = netflix_1.replace({np.nan: None})
netflix_1 = netflix_1.where(pd.notnull(netflix_1), None)

# Parsing every date_added value as a date. If one is invalid, just make it missing instead of throwing an error.
netflix_1["date_added"] = pd.to_datetime(
    netflix_1["date_added"],
    format="%d-%b-%y",
    errors="coerce"
)

# Renaming the 'rating' column to 'maturity_rating' to make it more clear
netflix_1 = netflix_1.rename(columns={"rating": "maturity_rating"})

# Constraining the date added to the pandemic period [2018,2021]
netflix_1 = netflix_1[
    (netflix_1["date_added"].dt.year >= 2018) &
    (netflix_1["date_added"].dt.year <= 2021)
]

# Convert duration into a number
netflix_1["duration"] = pd.to_numeric(netflix_1["duration"], errors="coerce")

# Replace missing values again because after converting dates and numbers, pandas may have created new missing values like NaT and NaN
netflix_1 = netflix_1.where(pd.notnull(netflix_1), None)
```

```python
print("There are", len(netflix_1), "rows in this dataset after cleaning.")
```

## Rotten Tomatoes Movies Dataset

```python
rotten_tomatoes_movies.head()
```

```python
print("There are", len(rotten_tomatoes_movies), "rows in this dataset before cleaning.")
```

```python
# Rename rating to maturity_rating for clarification
rotten_tomatoes_movies = rotten_tomatoes_movies.rename(columns={"rating": "maturity_rating"})

# Drop rows with missing maturity rating
rotten_tomatoes_movies = rotten_tomatoes_movies.dropna(subset=["maturity_rating"])

# Drop rows with missing audienceScore
rotten_tomatoes_movies = rotten_tomatoes_movies.dropna(subset=["audienceScore"])

# Drop rows with missing tomatoMeter
rotten_tomatoes_movies = rotten_tomatoes_movies.dropna(subset=["tomatoMeter"])

# Standardizing the maturity ratings so it matches the Netflix and Disney+ datasets (including a dash between letters, i.e. changing TVPG to TV-PG)
rating_map = {
    "TVPG": "TV-PG",
    "TV14": "TV-14",
    "TVMA": "TV-MA",
    "TVG": "TV-G",
    "TVY7": "TV-Y7"
}

rotten_tomatoes_movies["maturity_rating"] = (
    rotten_tomatoes_movies["maturity_rating"]
    .replace(rating_map)
)

# Convert releasedateStreaming to datetime
rotten_tomatoes_movies["releaseDateStreaming"] = pd.to_datetime(
    rotten_tomatoes_movies["releaseDateStreaming"],
    errors="coerce"
)

# Keep only 2018-2021 for releasedateStreaming
rotten_tomatoes_movies = rotten_tomatoes_movies[
    rotten_tomatoes_movies["releaseDateStreaming"].dt.year.between(2018, 2021)
]
```

```python
print("There are", len(rotten_tomatoes_movies), "rows in this dataset after cleaning.")
```

## IMDB Title Basics Dataset

```python
imdb_title_ratings = pd.read_csv("title.ratings.tsv", sep="\t")
imdb_title_basics = pd.read_csv("title.basics.txt", sep="\t")
```

```python
imdb_title_basics.head()
```

```python
print("There are", len(imdb_title_basics), "rows in this dataset before cleaning.")
```

```python
# Python is currently storing startYear as text, convert to numeric
imdb_title_basics["startYear"] = pd.to_numeric(
    imdb_title_basics["startYear"],
    errors="coerce"
)

# Filter to be within pandemic period [2018,2021]
imdb_title_basics = imdb_title_basics[
    imdb_title_basics["startYear"].between(2018, 2021)
]
```

```python
print("There are", len(imdb_title_basics), "rows in this dataset after cleaning.")
```

## IMDB Title Ratings Dataset

```python
imdb_title_ratings.head()
```

```python
print("There are", len(imdb_title_ratings), "rows in this dataset before cleaning.")
```

```python
# Only keeping the tconst that is also in the imdb_title_basics dataset. Otherwise, we would have natural left joined in SQL and it automatically would have 
# kept only the rows where the tconst is the same, but reading in the raw dataset would have exceeded the quota in Oracle.
imdb_title_ratings = imdb_title_ratings[
    imdb_title_ratings["tconst"].isin(imdb_title_basics["tconst"])
]
```

```python
print("There are", len(imdb_title_ratings), "rows in this dataset after cleaning.")
```
