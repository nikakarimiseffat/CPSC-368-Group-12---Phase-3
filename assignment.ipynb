## Question 1
What is the impact of the COVID-19 pandemic on changing audience/viewer sentiment towards all movies measured by average IMDb rating? 
### Hypothesis 
There is a statistically significant increase in the average IMDb rating for movies and TV shows released between the height of COVID-19 (2020, 2021) and 2018-2019, likely due to increased consumption and greater access to entertainment, with algorithms curated to users' tastes. \
Let the average imdb score in year (2020, 2021) to be $\mu$ and the average imdb score in year (2018, 2019) to be $\mu_0$.\
$H_0: \mu = \mu_0, H_A : \mu > \mu_0$.


```python
cur.execute("""
            DROP VIEW imdb_movies_ratings
            """)
```

```python
cur.execute("""
            CREATE VIEW imdb_movies_ratings AS
            SELECT 
            b.tconst, 
            b.primaryTitle, 
            b.startYear, 
            b.genres,
            r.average_ratings, 
            r.numVotes
            FROM imdb_title_basics b
            INNER JOIN imdb_ratings r ON b.tconst = r.tconst
            """)
```

```python
avg_rating_df = pd.read_sql("""
    SELECT AVG(average_ratings), STDDEV(average_ratings), SUM(numVotes)
    FROM imdb_movies_ratings
    WHERE startYear IN (2018, 2019)
""", connection)

print(avg_rating_df)
```

```python
avg_rating_df_muA = pd.read_sql("""
    SELECT AVG(average_ratings), STDDEV(average_ratings), SUM(numVotes)
    FROM imdb_movies_ratings
    WHERE startYear IN (2020, 2021)
""", connection)
print(avg_rating_df_muA)
```

## Question 2

### Hypothesis
The gap between Rotten Tomatoes’ Tomatometer (Critics) and Popcornmeter (Audience) scores widened significantly during the pandemic (2020-2021) compared to the pre-pandemic baseline, as critics focused on quality, while the general audience consumed more content to pass the time. 


We define $p$ as the proportion of films where the critic liked the movie more than the audience (meaning $score\_gap = Tomatometer - Popcornmeter > 0$).

Then the hypothesis is that: $H_0: p_{pandemic} \leq p_{pre}$, and $H_a: p_{pandemic} > p_{pre}$ CHANGE IT IF WRONG

```python
cur.execute("""
            DROP VIEW rt_with_gap
            """)
```

```python
cur.execute("""
    CREATE VIEW rt_with_gap AS
SELECT 
    id, 
    m.title, 
    audienceScore, 
    tomatoMeter, 
    (tomatoMeter - audienceScore) AS score_gap,
    release_year
FROM rotten_tomatoes rt INNER JOIN all_media_titles m ON rt.title = m. title
""")
```

```python
num_success_pan = pd.read_sql("""
    SELECT COUNT(score_gap)
    FROM rt_with_gap
    WHERE release_year IN (2020, 2021)
    AND score_gap >0
""", connection)
```

```python
num_success_pan
```

```python
total_num_pan = pd.read_sql("""
    SELECT COUNT(score_gap)
    FROM rt_with_gap
    WHERE  release_year IN (2020, 2021)
""", connection)
total_num_pan
```

```python
num_success_pan/total_num_pan
```

```python
num_success_pre = pd.read_sql("""
    SELECT COUNT(score_gap)
    FROM rt_with_gap
    WHERE release_year IN (2018,2019)
    AND score_gap >0
""", connection)
```

```python
num_success_pre
```

```python
total_num_pre = pd.read_sql("""
    SELECT COUNT(score_gap)
    FROM rt_with_gap
    WHERE release_year IN (2018,2019)
""", connection)
total_num_pre
```

We then perform hypothesis testing by a difference in proportion test. Firstly, we set the significance level $\alpha = 5\%$ 

```python
num_success_pre/total_num_pre
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
count = [num_success_pan, num_success_pre]
nobs = [total_num_pan, total_num_pre]

z_stat, p_value = proportions_ztest(count, nobs, alternative='larger')

print(z_stat)
print(p_value)
```

### Hypothesis 
During the pandemic, mature-rated media (R and TV-MA) showed an increase in total vote volume and a more consistent average rating than Parental-Guidance-rated media (PG-13, TV-PG, and TV-14). This divergence is primarily attributable to the transition in entertainment consumption from conventional cinema to digital streaming platforms (e.g., Netflix, Disney+).


```python
cur.execute("""
   DROP VIEW imdb_net_disney
""")
```

```python
cur.execute("""
    CREATE VIEW imdb_net_disney AS
 SELECT * FROM imdb_movies_ratings i
   INNER JOIN all_media_titles a 
   ON i.primaryTitle = a.title
   AND i.startYear = a.release_year
""")
```

```python
post_df = pd.read_sql("""
    WITH CategorizedMovies AS (
    SELECT imdb_net_disney.*,
           CASE 
               WHEN matuarity_rating IN ('TV-14', 'PG-13', 'TV-PG') THEN 'PG'
               WHEN matuarity_rating IN ('R', 'TV-MA') THEN 'Restricted'
           END AS rating_category
    FROM imdb_net_disney
)
SELECT rating_category, COUNT(*),AVG(AVERAGE_RATINGS), SUM(NUMVOTES), STDDEV(AVERAGE_RATINGS)
FROM CategorizedMovies
WHERE release_year IN (2020,2021)
GROUP BY rating_category
""", connection)
post_df
```
